```python
# Установка необходимых библиотек
!pip install pandas numpy matplotlib seaborn scikit-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score, classification_report
import warnings
warnings.filterwarnings('ignore')

print("=== ЛАБОРАТОРНАЯ РАБОТА №4 ===")
print("Вариант 25: Ozone Level Detection Data Set")
print()

# 1. ЗАГРУЗКА ДАННЫХ
print("=" * 80)
print("1. ЗАГРУЗКА ДАННЫХ")
print("=" * 80)

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/ozone/eighthr.data"
column_names = [f'feature_{i}' for i in range(1, 73)] + ['class']
data = pd.read_csv(url, header=None, names=column_names, na_values='?')

print(f"Размер исходного датасета: {data.shape}")

selected_columns = ['feature_27', 'feature_53', 'feature_71', 'class']
selected_data = data[selected_columns].copy()
selected_data.columns = ['WSR_AV', 'T_AV', 'SLP', 'class']

print(f"Выбранные признаки: WSR_AV, T_AV, SLP, class")
print(f"Первые 5 строк данных:")
print(selected_data.head())

# 2. ОБРАБОТКА ПРОПУЩЕННЫХ ЗНАЧЕНИЙ
print("\n" + "=" * 80)
print("2. ОБРАБОТКА ПРОПУЩЕННЫХ ЗНАЧЕНИЙ")
print("=" * 80)

print("Пропущенные значения до обработки:")
print(selected_data.isnull().sum())

selected_data = selected_data.dropna(subset=['class'])
selected_data['class'] = selected_data['class'].astype(int)

for feature in ['WSR_AV', 'T_AV', 'SLP']:
    missing_mask = selected_data[feature].isnull()
    if missing_mask.any():
        for class_val in selected_data['class'].unique():
            class_mask = selected_data['class'] == class_val
            missing_in_class = missing_mask & class_mask
            if missing_in_class.any():
                median_val = selected_data.loc[class_mask, feature].median()
                selected_data.loc[missing_in_class, feature] = median_val

print(f"Пропущенные значения после обработки:")
print(selected_data.isnull().sum())
print(f"Размер данных после обработки: {selected_data.shape}")

# 3. ДИАГРАММА РАССЕЯНИЯ С ПОДДИАПАЗОНАМИ
print("\n" + "=" * 80)
print("3. ДИАГРАММА РАССЕЯНИЯ С ПОДДИАПАЗОНАМИ")
print("=" * 80)

slp_min = selected_data['SLP'].min()
slp_max = selected_data['SLP'].max()
print(f"Минимальное значение SLP: {slp_min:.2f}")
print(f"Максимальное значение SLP: {slp_max:.2f}")

slp_bins = np.linspace(slp_min, slp_max, 11)
selected_data['SLP_bin'] = pd.cut(selected_data['SLP'], bins=slp_bins, labels=range(10))

print(f"Границы поддиапазонов SLP:")
for i in range(len(slp_bins)-1):
    print(f"  Поддиапазон {i}: [{slp_bins[i]:.2f}, {slp_bins[i+1]:.2f})")

plt.figure(figsize=(15, 10))
marker_sizes = [20 + i*15 for i in range(10)]
colors = ['red', 'blue']
class_labels = ['Class 0', 'Class 1']

for class_val in sorted(selected_data['class'].unique()):
    class_data = selected_data[selected_data['class'] == class_val]
    for bin_val in range(10):
        bin_data = class_data[class_data['SLP_bin'] == bin_val]
        if len(bin_data) > 0:
            plt.scatter(bin_data['WSR_AV'], bin_data['T_AV'],
                       s=marker_sizes[bin_val],
                       c=colors[int(class_val)],
                       alpha=0.6,
                       label=f'Class {class_val}' if bin_val == 0 else "")

plt.xlabel('WSR_AV', fontsize=12)
plt.ylabel('T_AV', fontsize=12)
plt.title('Диаграмма рассеяния: WSR_AV vs T_AV\n(размер маркера зависит от поддиапазона SLP)', fontsize=14)
plt.grid(True, alpha=0.3)

legend_elements_class = [plt.Line2D([0], [0], marker='o', color='w',
                          markerfacecolor=colors[i], markersize=8, label=class_labels[i])
                         for i in range(len(colors))]
legend1 = plt.legend(handles=legend_elements_class, title='Классы', loc='upper right')

legend_elements_size = [plt.Line2D([0], [0], marker='o', color='w',
                         markerfacecolor='gray', markersize=np.sqrt(marker_sizes[i])/2,
                         label=f'SLP bin {i}')
                        for i in range(0, 10, 2)]
plt.gca().add_artist(legend1)
plt.legend(handles=legend_elements_size, title='Поддиапазоны SLP', loc='lower right')

plt.tight_layout()
plt.show()

# 4. ЭМПИРИЧЕСКИЕ ФУНКЦИИ РАСПРЕДЕЛЕНИЯ (CDF)
print("\n" + "=" * 80)
print("4. ЭМПИРИЧЕСКИЕ ФУНКЦИИ РАСПРЕДЕЛЕНИЯ (CDF)")
print("=" * 80)

plt.figure(figsize=(18, 5))
features = ['WSR_AV', 'T_AV', 'SLP']
feature_names = ['WSR_AV', 'T_AV', 'SLP']
colors = ['blue', 'green', 'red']
line_styles = ['-', '--', '-.']

for i, (feature, name) in enumerate(zip(features, feature_names)):
    plt.subplot(1, 3, i+1)
    sorted_data = np.sort(selected_data[feature])
    y_vals = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
    plt.plot(sorted_data, y_vals, color=colors[i], linestyle=line_styles[i], linewidth=2)
    plt.xlabel(name, fontsize=12)
    plt.ylabel('Вероятность', fontsize=12)
    plt.title(f'CDF для {name}', fontsize=14)
    plt.grid(True, alpha=0.3)
    mean_val = selected_data[feature].mean()
    median_val = selected_data[feature].median()
    plt.axvline(mean_val, color='orange', linestyle=':', alpha=0.8, label=f'Среднее: {mean_val:.2f}')
    plt.axvline(median_val, color='purple', linestyle=':', alpha=0.8, label=f'Медиана: {median_val:.2f}')
    plt.legend()

plt.tight_layout()
plt.show()

# 5. КОРРЕЛЯЦИЯ И МАРЖИНАЛЬНАЯ ГИСТОГРАММА
print("\n" + "=" * 80)
print("5. КОРРЕЛЯЦИЯ И МАРЖИНАЛЬНАЯ ГИСТОГРАММА")
print("=" * 80)

correlation_matrix = selected_data[['WSR_AV', 'T_AV', 'SLP']].corr()
print("Корреляционная матрица:")
print(correlation_matrix)

corr_values = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        feat1 = correlation_matrix.columns[i]
        feat2 = correlation_matrix.columns[j]
        corr_val = abs(correlation_matrix.iloc[i, j])
        corr_values.append((feat1, feat2, corr_val, correlation_matrix.iloc[i, j]))

corr_values_sorted = sorted(corr_values, key=lambda x: x[2], reverse=True)

print(f"Пары признаков по убыванию корреляции:")
for feat1, feat2, abs_corr, corr in corr_values_sorted:
    print(f"  {feat1} - {feat2}: {corr:.4f}")

most_correlated = corr_values_sorted[0]
print(f"Наиболее коррелированные признаки: {most_correlated[0]} и {most_correlated[1]}")
print(f"Коэффициент корреляции: {most_correlated[3]:.4f}")

fig = plt.figure(figsize=(12, 10))
gs = fig.add_gridspec(2, 2, width_ratios=[4, 1], height_ratios=[1, 4],
                      left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.05, hspace=0.05)

ax_scatter = fig.add_subplot(gs[1, 0])
ax_hist_x = fig.add_subplot(gs[0, 0], sharex=ax_scatter)
ax_hist_y = fig.add_subplot(gs[1, 1], sharey=ax_scatter)

colors = ['red', 'blue']
class_labels = ['Class 0', 'Class 1']

for class_val in sorted(selected_data['class'].unique()):
    class_data = selected_data[selected_data['class'] == class_val]
    ax_scatter.scatter(class_data[most_correlated[0]], class_data[most_correlated[1]],
                      c=colors[int(class_val)], alpha=0.6, label=class_labels[int(class_val)], s=50)

for class_val in sorted(selected_data['class'].unique()):
    class_data = selected_data[selected_data['class'] == class_val]
    ax_hist_x.hist(class_data[most_correlated[0]], bins=30, alpha=0.6,
                   color=colors[int(class_val)], density=True)

for class_val in sorted(selected_data['class'].unique()):
    class_data = selected_data[selected_data['class'] == class_val]
    ax_hist_y.hist(class_data[most_correlated[1]], bins=30, alpha=0.6,
                   color=colors[int(class_val)], density=True, orientation='horizontal')

ax_scatter.set_xlabel(most_correlated[0], fontsize=12)
ax_scatter.set_ylabel(most_correlated[1], fontsize=12)
ax_scatter.set_title(f'Маржинальная гистограмма: {most_correlated[0]} vs {most_correlated[1]}\n(Корреляция: {most_correlated[3]:.4f})', fontsize=14)
ax_scatter.legend()
ax_scatter.grid(True, alpha=0.3)

ax_hist_x.set_ylabel('Плотность')
ax_hist_y.set_xlabel('Плотность')
ax_hist_x.tick_params(axis="x", labelbottom=False)
ax_hist_y.tick_params(axis="y", labelleft=False)

min_x = selected_data[most_correlated[0]].min()
max_x = selected_data[most_correlated[0]].max()
min_y = selected_data[most_correlated[1]].min()
max_y = selected_data[most_correlated[1]].max()

min_x_point = selected_data[selected_data[most_correlated[0]] == min_x].iloc[0]
max_x_point = selected_data[selected_data[most_correlated[0]] == max_x].iloc[0]
min_y_point = selected_data[selected_data[most_correlated[1]] == min_y].iloc[0]
max_y_point = selected_data[selected_data[most_correlated[1]] == max_y].iloc[0]

ax_scatter.annotate(f'Min {most_correlated[0]}\n({min_x:.2f})',
                   xy=(min_x, min_x_point[most_correlated[1]]),
                   xytext=(min_x + 0.1*(max_x-min_x), min_x_point[most_correlated[1]]),
                   arrowprops=dict(arrowstyle='->', color='black'),
                   fontsize=10)

ax_scatter.annotate(f'Max {most_correlated[0]}\n({max_x:.2f})',
                   xy=(max_x, max_x_point[most_correlated[1]]),
                   xytext=(max_x - 0.3*(max_x-min_x), max_x_point[most_correlated[1]]),
                   arrowprops=dict(arrowstyle='->', color='black'),
                   fontsize=10)

ax_scatter.annotate(f'Min {most_correlated[1]}\n({min_y:.2f})',
                   xy=(min_y_point[most_correlated[0]], min_y),
                   xytext=(min_y_point[most_correlated[0]], min_y + 0.1*(max_y-min_y)),
                   arrowprops=dict(arrowstyle='->', color='black'),
                   fontsize=10)

ax_scatter.annotate(f'Max {most_correlated[1]}\n({max_y:.2f})',
                   xy=(max_y_point[most_correlated[0]], max_y),
                   xytext=(max_y_point[most_correlated[0]], max_y - 0.2*(max_y-min_y)),
                   arrowprops=dict(arrowstyle='->', color='black'),
                   fontsize=10)

plt.tight_layout()
plt.show()

print(f"Экстремальные значения:")
print(f"  {most_correlated[0]}: min={min_x:.2f}, max={max_x:.2f}")
print(f"  {most_correlated[1]}: min={min_y:.2f}, max={max_y:.2f}")

# ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ
print("\n" + "=" * 80)
print("ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ")
print("=" * 80)

class_distribution = selected_data['class'].value_counts()
print("Распределение классов:")
print(class_distribution)

print(f"Доля классов:")
for class_val, count in class_distribution.items():
    percentage = (count / len(selected_data)) * 100
    print(f"  Класс {class_val}: {count} записей ({percentage:.1f}%)")

print(f"Общая статистика:")
print(f"  Всего записей: {len(selected_data)}")
print(f"  Количество признаков: 3")
print(f"  Количество классов: {len(class_distribution)}")

selected_data.to_csv('processed_ozone_data.csv', index=False)
print(f"Обработанные данные сохранены в файл: processed_ozone_data.csv")

print("\n" + "=" * 80)
print("ЛАБОРАТОРНАЯ РАБОТА ЗАВЕРШЕНА")
print("=" * 80)
```
